```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pit_data = pd.read_excel('Better_Data/NC Homeless Data.xlsx')
```

```python
pit_data
```

```python
x_axis = pit_data["Year"]
y_axis = pit_data["Overall Homeless"]
plt.plot(x_axis, y_axis)
plt.title("Overall Homelessness in NC 2013-2019")
plt.xlabel(" Year ")
plt.ylabel("'Point in Time' Count")
plt.savefig("../Images/overallhomeless.png")
```
