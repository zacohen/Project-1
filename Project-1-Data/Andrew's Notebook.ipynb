```python
#imports
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from scipy.stats import linregress
```

```python
#Import data
Data = pd.read_csv("Resources/PovertyGender.csv")
Data.head()
```

```python
#Save age groups in variable
eighteen_to_twentyfour = Data.loc[Data["Age"] == "18 to 24 Years"]
twentyfive_to_thirtyfour = Data.loc[Data["Age"] == "25 to 34 Years"]
thirtyfive_to_fourtyfour = Data.loc[Data["Age"] == "35 to 44 Years"]
fourtyfive_to_fiftyfour = Data.loc[Data["Age"] == "45 to 54 Years"]
fiftyfive_to_sixtyfour = Data.loc[Data["Age"] == "55 to 64 Years"]
sixtyfive_to_seventyfour = Data.loc[Data["Age"] == "65 to 74 Years"]
seventyfive_and_up = Data.loc[Data["Age"] == "75 Years & Over"]
```

```python
#18 to 24 Graph

#plt.scatter(eighteen_to_twentyfour["Year"], eighteen_to_twentyfour["Poverty Population"])
#plt.title("18 to 24 year olds")
#plt.xlabel("Year")
#plt.ylabel("Number of people below the poverty line")

x_values = eighteen_to_twentyfour["Year"]
y_values = eighteen_to_twentyfour["Poverty Population"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(2014,9000),fontsize=10,color="red")
plt.xlabel('Year')
plt.ylabel('Number of people below the poverty line')
plt.title("18 to 24")
print(f" R-Value = {rvalue}")
print(f" P-Value = {pvalue}")
print(f" Standard Deviation = {stderr}")
plt.show()
```
